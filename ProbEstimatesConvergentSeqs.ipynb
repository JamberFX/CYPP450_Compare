```python

StructureTreeFile = "C:/Users/jamie/OneDrive/Documents/Biology Projects/CYP450 Manuscript for GBE/DataSets/1630 CYPP450s on both AlphaFold and Uniprot(Reviewed) Seq_Struct_Analysis/Trees/FT_of_1630_Reviewed_UniProt_CYPP450s.nwk"
FastaFileLocation= "C:/Users/jamie/OneDrive/Documents/Biology Projects/CYP450 Manuscript for GBE/DataSets/1630 CYPP450s on both AlphaFold and Uniprot(Reviewed) Seq_Struct_Analysis/Sequences/CYP450AFSeqs1630.fasta"


from Bio import Phylo, SeqIO, pairwise2
from Bio.Align import substitution_matrices
import numpy as np
import pandas as pd
import PCDTW
from IPython.display import clear_output


# Load Newick tree
tree = Phylo.read(StructureTreeFile, "newick")
leaf_names = [leaf.name for leaf in tree.get_terminals()]

# Load sequences into a dictionary keyed by leaf names
seq_records = list(SeqIO.parse(FastaFileLocation, "fasta"))
namerlst=[]
sequencenamerlst=[]
for i in range(len(seq_records)):
    namerlst.append(seq_records[i].description)
    sequencenamerlst.append(seq_records[i].seq)

seq_dict = dict(zip(namerlst, sequencenamerlst))

# Fetch sequence by leaf name
def get_seq(name):
    return str(seq_dict[name])

# Find closest neighbor (based on tree distance)
def get_closest_neighbor(tree, target_leaf):
    distances = {
        leaf.name: tree.distance(target_leaf, leaf.name)
        for leaf in tree.get_terminals() if leaf.name != target_leaf
    }
    return min(distances, key=distances.get)

#Calculate identity with blosum 45 matrix
def blosum45_identity(seq1, seq2):
    matchcount=0
    matrix = substitution_matrices.load("BLOSUM45")
    GapOpen=-3
    GapExtend=-1
    alignment=pairwise2.align.globalds(seq1,seq2,matrix,GapOpen,GapExtend)
    alignedseqA=alignment[0].seqA
    alignedseqB=alignment[0].seqB
    for i in range(len(alignedseqA)):
        if alignedseqA[i]==alignedseqB[i]:
            matchcount+=1
    identity=round((matchcount / len(alignedseqA))*100)

    return identity


# Example: run on all leaves and print results
leafnameslst=[]
identitylst=[]
pcdtwdistlst=[]
simtypelst=[]
seqlst=[]
testDataDF=pd.DataFrame()
counter=0
identity=0
# for leaf in leaf_names:
for i in range(len(leaf_names)):
# for i in range(10):
    leaf=leaf_names[i]
    clear_output(wait=True)  # Clears previous output

    print(counter)
    
    
    counter+=1
    try:
        neighbor = get_closest_neighbor(tree, leaf)
        
        seq1, seq2 = get_seq(leaf), get_seq(neighbor)

        identity = blosum45_identity(seq1, seq2)
        print(identity)
        pcdtw_dist = PCDTW.PCDTWDist(seq1, seq2)

        leafnameslst.append(leaf)
        identitylst.append(identity)
        pcdtwdistlst.append(pcdtw_dist)
        simtypelst.append("TestData")
        seqlst=("Not_Needed")

    except KeyError as e:
        leafnameslst.append("Error")
        identitylst.append("Error")
        pcdtwdistlst.append("Error")
        simtypelst.append("TestData")
        seqlst=("Not_Needed")

testDataDF["UNIPROTID"]=leafnameslst
testDataDF["SIMTYPE"]=simtypelst
testDataDF["SEQUENCE"]=seqlst
testDataDF["IDENTITY"]=identitylst
testDataDF["UNNORMDTWPERRESIDUE"]=pcdtwdistlst

display(testDataDF)

```

```python

# Join the two dataframes
combined_df = pd.concat([SmallSimDF, testDataDF], ignore_index=True)
display(combined_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, accuracy_score, precision_score, f1_score

# ---- Step 1: Normalize Entire Dataset ----
features = ['IDENTITY', 'UNNORMDTWPERRESIDUE']
scaler = StandardScaler()
df_scaled = combined_df.copy()
df_scaled[features] = scaler.fit_transform(combined_df[features])  # Normalizes all SIMTYPEs

# ---- Step 2: Filter to Directed and Random for training ----
train_df = df_scaled[df_scaled['SIMTYPE'].isin(['Directed', 'Random'])].copy()

# ---- Step 3: Prepare X and y ----
X = train_df[features].values
y = train_df['SIMTYPE'].map({'Directed': 0, 'Random': 1}).values  # Binary encoding

# ---- Step 4: Set up cross-validation ----
model = LogisticRegression(max_iter=1000)
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# ---- Step 5: Evaluate performance ----
scores = {
    'Accuracy': cross_val_score(model, X, y, cv=cv, scoring='accuracy'),
    'Precision': cross_val_score(model, X, y, cv=cv, scoring='precision'),
    'F1': cross_val_score(model, X, y, cv=cv, scoring='f1')
}

# ---- Step 6: Report scores ----
for metric, values in scores.items():
    print(f"{metric}: {values.mean():.3f} ± {values.std():.3f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score

# ---- Step 1: Normalize Entire Dataset ----
features = ['IDENTITY', 'UNNORMDTWPERRESIDUE']
scaler = StandardScaler()
df_scaled = combined_df.copy()
df_scaled[features] = scaler.fit_transform(df_scaled[features])  # Normalizes ALL SIMTYPEs

# ---- Step 2: Filter to Directed and Random for training ----
train_df = df_scaled[df_scaled['SIMTYPE'].isin(['Directed', 'Random'])].copy()

# ---- Step 3: Prepare features (X) and binary labels (y) ----
X = train_df[features].values
y = train_df['SIMTYPE'].map({'Directed': 0, 'Random': 1}).values

# ---- Step 4: Set up SVM model with RBF kernel and cross-validation ----
model = SVC(kernel='rbf', probability=True, random_state=42)
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# ---- Step 5: Evaluate metrics ----
scores = {
    'Accuracy': cross_val_score(model, X, y, cv=cv, scoring='accuracy'),
    'Precision': cross_val_score(model, X, y, cv=cv, scoring='precision'),
    'F1': cross_val_score(model, X, y, cv=cv, scoring='f1')
}

# ---- Step 6: Display results ----
for metric, values in scores.items():
    print(f"{metric}: {values.mean():.3f} ± {values.std():.3f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import joblib

# ---- Step 1: Normalize Entire Dataset ----
features = ['IDENTITY', 'UNNORMDTWPERRESIDUE']
df_scaled = combined_df.copy()

scaler = StandardScaler()
df_scaled[features] = scaler.fit_transform(df_scaled[features])

# ---- Step 2: Save the Scaler ----
joblib.dump(scaler, 'standard_scaler.pkl')

# ---- Step 3: Filter Training Data (Directed + Random) ----
train_df = df_scaled[df_scaled['SIMTYPE'].isin(['Directed', 'Random'])].copy()
X_train = train_df[features].values
y_train = train_df['SIMTYPE'].map({'Directed': 0, 'Random': 1}).values

# ---- Step 4: Train Final SVC Model (RBF Kernel) ----
model = SVC(kernel='rbf', probability=True, random_state=42)
model.fit(X_train, y_train)

# ---- Step 5: Save the Trained Model ----
joblib.dump(model, 'svc_rbf_final_model.pkl')

# ---- Step 6: Evaluate TestData ----
test_df = df_scaled[df_scaled['SIMTYPE'] == 'TestData'].copy()
X_test = test_df[features].values

# Predict probability and class
probs = model.predict_proba(X_test)[:, 1]  # Probability of class "Random"
preds = model.predict(X_test)              # 0 = Directed, 1 = Random

# ---- Step 7: Add Predictions to TestData ----
test_df['RandomProbability'] = probs
test_df['PredictedClass'] = preds
test_df['PredictedSIMTYPE'] = test_df['PredictedClass'].map({0: 'Directed', 1: 'Random'})

# ---- Step 8: Output Summary ----
result_df = test_df[['UNIPROTID', 'IDENTITY', 'UNNORMDTWPERRESIDUE', 'RandomProbability', 'PredictedSIMTYPE']]
print(result_df.head())
```
